```python
%matplotlib inline
import numpy as np
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt
```

```python
fig, (ax01, ax02, ax03) = plt.subplots(1, 3, figsize=(18, 4),)

ax01.plot(np.linspace(0, 0.3, 300),
         np.zeros(300), color='C0')
ax01.plot([0.3, 0.3],
          [0, 1 / 0.3], color='C0')
ax01.plot(np.linspace(0.3, .6, 250), 
         stats.beta.pdf(np.linspace(0, 1, 250), 1, 1) / 0.3 )
ax01.plot([0.6, 0.6],
          [0, 1 / 0.3], color='C0')
ax01.plot(np.linspace(0.6, 1.0, 400),
         np.zeros(400), color='C0')

ax02.plot(np.linspace(0, 0.3, 300),
         np.zeros(300), color='C0')
ax02.plot([0.3, 0.3],
          [0, 9 / 0.3], color='C0')
ax02.plot(np.linspace(0.3, 0.6, 250),
         stats.beta.pdf(np.linspace(0, 1, 250), 0.25, 0.25) / 0.3)
ax02.plot([0.6, 0.6],
          [0, 9 / 0.3], color='C0')
ax02.plot(np.linspace(0.6, 1.0, 400),
         np.zeros(400), color='C0')


ax03.plot([0.3, 0.3],[0, 0.5])
ax03.plot(np.array([1, 1]) * 0.3 + 0.3, 
          [0, 0.5], color='C0')
ax03.scatter(np.array([0, 1]) * 0.3 + 0.3,
             [0.5, 0.5], color='C0')
ax03.plot(np.linspace(0, 1.0, 1000),
         np.zeros(1000), color='C0')

height = 5.1
ax01.set_ylim(-0.1, height)
ax02.set_ylim(-0.1, height)
ax03.set_ylim(-0.1, height)
ax01.set_xlim(-0.05, 1.05)
ax02.set_xlim(-0.05, 1.05)
ax03.set_xlim(-0.05, 1.05)
ax01.set_title('Denisty funtion of 0.3 Uniform(0, 1) + 0.3')
ax02.set_title('Denisty function of 0.3 Beta(0.25, 0.25) + 0.3')
ax03.set_title('Mass funtion of 0.3 Bernoulli(0.5) + 0.3');
fig.savefig('../figures/graphs_rdpg_distributions_visualization.png',
             bbox_inches='tight',
             dpi=300)
```
